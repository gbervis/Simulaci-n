# Modelo general de inventario
El problema del inventario tiene que ver con guardar en reserva un artículo para satisfacer las fluctuaciones de la demanda. 
El exceso de existencias de un artículo aumenta el costo del capital y de almacenamiento, y la escasez de existencias interrumpe la producción o las ventas. El resultado es buscar un nivel de inventario que balancee las dos situaciones extremas minimizando una función de costo apropiada. 
El problema se reduce a controlar el nivel del inventario diseñando una política de inventario que responda dos preguntas:
1. ¿Cuánto pedir?
2. ¿Cuándo pedir?

Un sistema de inventario puede requerir 
revisiones periódicas (por ejemplo, pedir al inicio de cada semana o cada mes). 
Alternativamente, el sistema puede estar basado en 
revisiones continuas, colocando un nuevo pedido cuando el nivel del inventario se reduce a un punto de volver a pedir específico. 

Un ejemplo de los dos tipos ocurre en las pulperías. 
La revisión es periódica si el artículo se repone cada semana o cada mes. 
La revisión es continua si la reposición ocurre siempre que el nivel del inventario se reduce por debajo de un determinado nivel.

# Demanda en los modelos de inventario
En general, la complejidad de los modelos de inventario depende de si la demanda es determinística o probabilística. Dentro de ambas categorías, la demanda puede variar, o no, con el tiempo.

En situaciones prácticas, el patrón de la demanda en un modelo de inventario puede asumir uno de cuatro tipos: 
1. Determinístico y constante (estático) con el tiempo. 
2. Determinístico y variable (dinámico) con el tiempo. 
3. Probabilístico y estacionario a lo largo del tiempo. 
4. Probabilístico y no estacionario a lo largo del tiempo. 

# Demanda en los modelos de inventario
1. **Determinístico y constante (estático) con el tiempo.** Si la demanda mensual promedio (registrada a lo largo de varios años) es “de manera aproximada” constante y la variación es razonablemente pequeño (< 20%), entonces la demanda puede considerarse determinística y constante.
2. **Determinístico y variable (dinámico) con el tiempo.** Si la demanda mensual promedio varía de manera apreciable entre los diferentes meses pero la variación permanece razonablemente pequeño en todos los meses, entonces la demanda puede considerarse determinística pero variable.
3. **Probabilístico y estacionario a lo largo del tiempo.** Si en el caso 1 la variación es alta (> 20%) pero aproximadamente constante, entonces la demanda es probabilística y estacionaria. 
4. **Probabilístico y no estacionario a lo largo del tiempo.** El caso restante es la demanda probabilística no estacionaria, la cual ocurre cuando los promedios y los coeficientes de variación varían apreciablemente mes con mes.

# MODELOS ESTÁTICOS DE CANTIDAD DE PEDIDO ECONÓMICO (EOQ)

# Modelo EOQ clásico
El más simple de los modelos de inventario implica una demanda de tasa constante con reposición de pedidos instantánea y sin escasez. Defina
$𝑦$ : Cantidad de pedido (número de unidades)
$𝐷$ : Tasa de demanda (unidades por unidad de tiempo)
$𝑡_0$  : Duración del ciclo de pedido (unidades de tiempo)
Cuando el inventario llega al nivel cero, se recibe al instante un pedido de $𝑦$ unidades de tamaño. Las existencias se agotan uniformemente a una tasa de demanda constante, $𝐷$. El ciclo de pedido de este patrón es
$𝑡_0= \displaystyle\frac{𝑦}{𝐷}$  unidades de tiempo.



El modelo de costo requiere dos parámetros de costo.
$𝐾$: Costo de preparación asociado con la colocación de un pedido (dólares por pedido)
$ℎ$: Costo de retención (dólares por unidad de inventario por unidad de tiempo)

Dado que el nivel de inventario promedio es $𝑦/2$, el costo total por unidad de tiempo (TCU, por sus siglas en inglés) es
$$𝑇𝐶𝑈(𝑦) = \frac{𝐾}{𝑡_0} + ℎ( \frac{𝑦}{2})$$

La política de inventario óptima para el modelo propuesto es

**Pedido de $𝒚^∗ = \sqrt{\displaystyle\frac{𝟐𝑲𝑫}{𝒉}}$ unidades cada $𝒕_𝒐^∗= \displaystyle \frac{𝒚^∗}{𝑫}$ unidades de tiempo.**


Un nuevo pedido no tiene que recibirse en el instante que se pide. En su lugar, puede ocurrir un tiempo de espera (tiempo de anticipación) positivo $𝐿$, entre la colocación y el recibo de un pedido como se muestra en la figura. En este caso el punto de volver a pedir (punto de reorden) ocurre cuando el nivel del inventario se reduce a $𝐿𝐷$ unidades.

La figura asume que el tiempo de espera $𝐿$ es menor que la duración del ciclo, así se define el tiempo de espera efectivo como 
$$𝐿_𝑒=𝐿−𝑛∗𝑡_0^∗$$
El parámetro $𝑛$ es el valor entero más grande no mayor que $𝐿/𝒕_𝟎^∗$. Y así, la política de inventario puede volverse a formular como 

**Pedir la cantidad $𝒚^∗$ siempre que el nivel del inventario se reduzca en $𝑳_𝒆 𝑫$ unidades**

# Modelo EOQ probabilístico

Este modelo permite faltantes en la demanda, la política requiere ordenar la cantidad y siempre que el inventario caiga al nivel "𝑅". Como en el caso determinista, el nivel de reorden "𝑅" es una función del tiempo de entrega, entre colocar y recibir un pedido. Los valores óptimos de "𝑅", se determinan minimizando el costo esperado por unidad de tiempo que incluye la suma de los costos de preparación, conservación y faltante.

El modelo tiene 5 suposiciones
- La demanda no satisfecha durante el tiempo de entrega se acumula.
- No se permite más de una orden pendiente.
- La distribución de la demanda durante el tiempo de entrega permanece estacionaria (sin cambio) con el tiempo.
- A diferencia de los otros modelos, aquí ya no hay completa certidumbre sobre la demanda, es decir, no se sabe con certeza.
- En éste modelo se necesitan herramientas estadísticas para poder estimar el punto en donde se debe pedir para así no quedarse sin existencias con cierto nivel de confianza.

**Ejercicio:** La demanda de azúcar en una tienda sigue una distribución exponencial con media de 100 Kg por día. El dueño de la tienda revisa el inventario cada 7 días, y hace un pedido a la planta igual a la capacidad de la bodega menos la cantidad de azúcar que tiene disponible en ese momento. La entrega es inmediata. La capacidad de la bodega es de 700 Kg. El costo de ordenar es 1000 córdobas la orden. El costo del faltante es de 6 córdobas por Kg, y el costo por mantener el inventario es de un córdoba por Kg. Determine el costo total promedio por día para un horizonte de dos meses, así como el comportamiento del inventario. 
- Realice la gráfica de la primera simulación. 
- Realice 10000 simulaciones.

```python
n = 10000 #No. de simulaciones
SPCt = 0
A = []

for j in range(n):
    Ii = 700

    SCt = 0  # Suma costo total

    for i in range(60):
        I = Ii
        if(j==1):
            A = np.append(A,I)

        Demanda = round(rd.exponential(100),1)
        if(I >= Demanda):
            Ventas = Demanda
            CF = 0
        else:
            Ventas = I
            CF = 6*(Demanda - Ventas) #Costo por faltantes
        If = I - Ventas

        CO = 0  # Costo de Orden
        if(i % 7 == 0):
            CO = 1000
            If = 700
        CI = 1*(Ii+If)/2 #Costo por tener el inventario
        Ct = CF + CI + CO#Costo Total
        SCt = SCt + Ct
        Ii = round(If,1)#Inv final = Inv Inicial
    PCt = SCt / 60 #Promedio costo total diario
    SPCt = SPCt + PCt
PSPCt = SPCt / n
print(f"El promedio del costo total diario es {PSPCt}")

plt.plot(A)
plt.show()
```

**Ejercicio:** Las luces de neón en el campus de la Universidad de Arkansas se reemplazan siguiendo una distribución normal con media 100 y desviación estándar 10. La planta física pide las luces de neón de forma periódica. Iniciar un pedido de compra cuesta $\$100$. Se estima que el costo de una luz de neón almacenada es de aproximadamente $\$0.02$ por día. El tiempo de espera entre la colocación y la recepción de un pedido sigue una distribución de Poisson con media 12 días. Determine la política de inventario de revisión continua óptima para pedir las luces de neón y el costo de inventario diario asociado a la política propuesta usando un código de simulación. 

<FONT COLOR = 'RED'> Realice 100,000 simulaciones.</FONT>

```python
m = 100000 #no. de simulaciónes

K = 100 #Costo de preparación de un pedido
h = 0.02 #Costo de retención

LeT, yT, TCUT, t0= 0,0,0,0

for i in range(m):
    D = round(rd.normal(100,10), 0) #Demanda
    L = rd.poisson(12) #Tiempo de espera
    y = np.sqrt((2*K*D)/h) #Cantidad de pedido
    t_0 = y / D #Duración del ciclo
    n = L//t_0
    L_e = L - n*t_0 #Tiempo de espera efectivo
    TCU = K/t_0 + h*(y/2) #Costo Total por unidad de tiempo

    LeT = L_e * D + LeT
    yT = y + yT
    TCUT = TCUT + TCU
    t0 = t_0 + t0 #Para la duración promedio del ciclo

yf = round(yT/m, 0)
LeD = round(LeT / m, 0)
TCUTf = round(TCUT / m, 0)

print(f"Pedir {yf} unidades siempre que el inventario se reduzca en {LeD}")
print(f"El costo de inventario asociado con la política propuesta es: {TCUTf}")
print(f"La duración del ciclo asociado es {round(t0/m,0)}")
```
