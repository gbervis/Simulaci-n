# Aplicando el Método Monte Carlo para la estimación de Pi

**Tomado de Hands-On Simulation page 91**

El método Monte Carlo es una estrategia de resolución de problemas que usa la estadística. Si consideramos $P$ como la probabilidad de cierto evento, entonces podemos simular aleatoriamente este evento y obtener $P$ encontrando el cociente entre el número de veces que ha ocurrido nuestro evento y el número total de simulaciones:
$$ P = \frac{\text{número de veces que ha ocurrido el evento}}{\text{número total de simulaciones}} $$

Podemos aplicar esta estrategia para obtener una aproximación de Pi. Pi ($\pi$) es una constante matemática que indica la relación entre la longitud de una circunferencia y su diámetro. De hecho, si denotamos $C$ como la longitud de una circunferencia y $d$ como su diámetro, sabemos que $C = \pi*d$. La longitud de una circunferencia cuyo diámetro es igual a 1 es $\pi$.

**Nota importante:** Usualmente, aproximamos el valor de Pi con 3.14 para simplificar los cálculos. Sin embargo, $\pi$ es un número irracional, esto es que tiene infinitos decimales que no tienen un patrón de repetición.

Dado un círculo de radio 1, este puede ser inscrito en un cuadrado de longitud 2. Por conveniencia, consideraremos una fracción del circulo, como se muestra a continuación:


Analizando la fígura anterior, podemos ver que el área del cuadrado en azul es 1 y que el área del sector circular en amarillo (1/4 del círculo) es pi/4. Podemos generar aleatoriamente un número grande de puntos dentro del cuadrado. **Si se genera un gran número de puntos podemos aproximar la medida del área contando el número de puntos contenidos en cada una de ellas. (Mejorar traducción)**

Si generamos $N$ números pseudoaleatorios dentro del cuadrado, el número de puntos que caen dentro de la región circular, denotado por $M$, dividido por el número total de números generados, $N$, tendremos que aproximar el área del sector circular y entonces habremos aproximado pi/4. De esto, podemos concluir la siguiente ecuación:
$$\pi = \frac{4*M}{N}$$

Entre más grande sea el número de puntos generados, mayor será la precisión de Pi que podemos obtener.

```python
# Librerias necesarias para el método
import math
import random
import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros
N = 100000
M = 0

# En esto guardaremos los puntos
XCircle=[]
YCircle=[]
XSquare=[]
YSquare=[]

for p in range(N):
    x=random.random()
    y=random.random()
    if(x**2+y**2 <= 1):
        M+=1
        XCircle.append(x)
        YCircle.append(y)
    else:
        XSquare.append(x)
        YSquare.append(y)
        
Pi = 4*M/N
print('N=%d M=%d Pi=%.7f' %(N,M,Pi))
```

```python
XLin=np.linspace(0,1)
YLin=[]
for x in XLin:
    YLin.append(math.sqrt(1-x**2))
    
plt.axis ('equal')
plt.grid (which='major')
plt.plot (XLin , YLin, color='red' , linewidth='4')
plt.scatter(XCircle, YCircle, color='yellow', marker
='.')
plt.scatter(XSquare, YSquare, color='blue' , marker
='.')
plt.title ('Método Monte Carlo para la estimación de Pi')
plt.show()
```
