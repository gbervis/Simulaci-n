La función $\texttt{numpy.random.choice(a, size = None, replace = True, p = None)}$ elige una muestra aleatoria de un arreglo 1-D.

**Parámetros:** Álgunos parámetros se omitirán.
- $\texttt{a}$ : nombre del arrreglo.
- $\texttt{size}$ : entero o tupla de enteros.
- $\texttt{replace}$ : Cuando la muestra es con reemplazamiento o no. Si se omite, es con reemplazamiento.
- $\texttt{p}$ : arreglo de probabilidades, asociadas a los elementos del arreglo $\texttt{a}$.
